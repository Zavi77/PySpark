```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('abc').getOrCreate()
```

```python
df = spark.read.csv('data1.csv',header=True,inferSchema=True)
```

```python
df.show(1)
```

## Drop na values

```python
df.na.drop(how='any',thresh=2,subset='artist').show(1)
```

## Fill na values

```python
df.na.fill('Not present',subset=['artist'])
```

## Imputer of null values

```python
from pyspark.ml.feature import Imputer

imp = Imputer(inputCol='energy').setStrategy('mean')
```

```python
imp.fit(df).transform(df).show(1)
```

## Filter data

```python
df.select('artist').show(5)
```

```python
df.filter((df['duration_ms'] > 300000) & (df['artist'] == 'Junior Boys')).show(5)
```

## Group by

```python
df.groupby('artist').mean('duration_ms').show(5)
```
