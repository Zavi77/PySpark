```python
from pyspark.sql import SparkSession
```

## Create Session

```python
spark = SparkSession.builder.appName('abc').getOrCreate()
```

```python
spark
```

## Get Data

```python
data = spark.read.csv('data.csv', header= True ,inferSchema=True)
```

```python
data.show(3)
```

```python
data.printSchema()
```

```python
data.columns
```

```python
data.select('unit_price').show(4)
```

## drop column

```python
data = data.drop('is_action','is_exclusive_to_us')
```

```python
data.columns
```

## Add column

```python
data.withColumn('new',data['unit_price']**2).show(3)
```

```python
data.withColumnRenamed('new','new price').show(5)
```
